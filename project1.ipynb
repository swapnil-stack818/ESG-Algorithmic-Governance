```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

def run_esg_visual_audit(file_path):
    print("==================================================================")
    print("STARTING INTERDISCIPLINARY ESG VISUAL AUDIT PIPELINE")
    print("==================================================================\n")

    # 1. LOAD DATASET
    try:
        df = pd.read_csv(file_path)
    except Exception as e:
        print(f"[ERROR] Failed to load dataset: {e}")
        return

    # 2. DATA CLEANING & PREPROCESSING
    features = ['Environment Risk Score', 'Governance Risk Score', 'Social Risk Score']
    df_clean = df.dropna(subset=features).copy()

    for feature in features:
        df_clean[feature] = pd.to_numeric(df_clean[feature], errors='coerce')
    df_clean = df_clean[(df_clean[features] >= 0).all(axis=1)]

    # 3. UNSUPERVISED MACHINE LEARNING (CLUSTERING)
    X = df_clean[features]
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
    df_clean['Behavioral_Cluster'] = kmeans.fit_predict(X_scaled)

    # 4. ANOMALY DETECTION (EUCLIDEAN DISTANCE)
    centroids = kmeans.transform(X_scaled)
    df_clean['Anomalous_Distance'] = np.min(centroids, axis=1)

    # Print tabular summary for validation
    print("--- TOP 5 ANOMALOUS CORPORATE PROFILES ---")
    audit_results = df_clean.sort_values(by='Anomalous_Distance', ascending=False).head(5)
    print(audit_results[['Name', 'Sector', 'Environment Risk Score', 'Governance Risk Score', 'Anomalous_Distance']].to_string(index=False))

    # ==========================================
    # GRAPHICAL ANALYSIS PIPELINE
    # ==========================================
    print("\n[SYSTEM] Generating graphical analytical charts...")

    # Set a clean, academic visual style
    sns.set_theme(style="whitegrid")

    # Plotting the data points across two major dimensions (Environment vs Governance)
    scatter = plt.scatter(
        df_clean['Environment Risk Score'],
        df_clean['Governance Risk Score'],
        c=df_clean['Behavioral_Cluster'],
        cmap='viridis',
        alpha=0.5,
        edgecolors='none',
        s=40
    )

    # Highlight and label the top outliers on the graph
    top_outliers = df_clean.sort_values(by='Anomalous_Distance', ascending=False).head(5)
    plt.scatter(
        top_outliers['Environment Risk Score'],
        top_outliers['Governance Risk Score'],
        color='#D63031',
        marker='X',
        s=130,
        edgecolors='black',
        label='Flagged Governance/Social Anomalies'
    )

    # Add clear, non-overlapping annotations for the flagged companies
    for i, txt in enumerate(top_outliers['Name']):
        # Shorten company name to the first two words to keep chart readable
        short_name = " ".join(txt.split()[:2])
        plt.annotate(
            short_name,
            (top_outliers['Environment Risk Score'].iloc[i], top_outliers['Governance Risk Score'].iloc[i]),
            textcoords="offset points",
            xytext=(10,-5),
            ha='left',
            fontsize=9,
            fontweight='bold',
            color='#2D3436',
            bbox=dict(boxstyle="round,pad=0.2", fc="yellow", alpha=0.3, ec="gray", lw=0.5)
        )

    # Formatting axes and labels cleanly
    plt.xlabel('Environmental Risk Index', fontsize=12, fontweight='semibold')
    plt.ylabel('Governance Risk Index', fontsize=12, fontweight='semibold')
    plt.title('Algorithmic ESG Audit: Clustering & Structural Anomaly Detection\n(S&P 500 Corporate Cohort)', fontsize=14, fontweight='bold', pad=15)
    plt.legend(loc='upper right', frameon=True, facecolor='white', edgecolor='none')

    # Prevent layout truncation
    plt.tight_layout()

    # Save the plot cleanly to your environment files
    plt.savefig('esg_risk_anomalies_chart.png', dpi=300)
    plt.close()
    print("[SYSTEM] High-resolution graphical chart saved successfully as 'esg_risk_anomalies_chart.png'!")

# Execute in your environment
run_esg_visual_audit("SP 500 ESG Risk Ratings.csv")
```

Project Title: Algorithmic Governance and Corporate Risk: An Unsupervised Machine Learning Framework for Automated ESG Data Auditing
Project Description: This independent, interdisciplinary project applies data analytics and unsupervised machine learning to evaluate corporate transparency and mitigate systemic risk within financial markets. The project bridges advanced data science (engineering) with corporate governance, environmental economics, and public trust (social/economic sciences).
The primary objective is to move beyond passive, paper-based compliance checks by building an automated, data-driven audit pipeline. It models standard behavioral archetypes among major corporations and dynamically isolates statistical anomalies—instances where a company’s disclosed environmental or social risk profiles are mathematically conflicting with its internal governance and structural compliance scores.
Data Source: The project utilizes an empiric, multi-dimensional dataset containing the ESG (Environmental, Social, and Governance) Risk Ratings for the S&P 500 Cohort, natively sourced from Sustainalytics' comprehensive corporate sustainability indexes. The data includes individual asset identifiers, sector categorizations, and granular, independent risk metrics mapped across environmental, social, and governance dimensions.
Methodology & Technical Execution: 1. Data Cleaning & Engineering Pipeline: Developed a Python-based preprocessing workflow that sanitizes categorical attributes, enforces strict numeric type-casting, and applies statistical standardization (StandardScaler) to equalize weight distributions across disparate risk metrics.
2.	Unsupervised Behavioral Clustering: Implemented a K-Means Clustering algorithm to automatically segment the S&P 500 cohort into multi-dimensional behavioral patterns representing systemic operational models.
3.	Statistical Anomaly Detection: Formulated a distance-based tracking mechanism that computes the Euclidean distance ($d$) of each data point to its closest cluster centroid. Entities exhibiting extreme spatial distances are mathematically flagged as structural outliers.
4.	Graphical Analytical Projection: Integrated a high-resolution data visualization pipeline using Matplotlib and Seaborn. The script automatically generates a multi-variable scatter plot mapping the industry clusters, overlays automated text-bounding boxes, and applies a visual heat-marker to pinpoint high-risk corporate anomalies for deeper regulatory scrutiny.
Interdisciplinary Societal & Economic Impact: By analyzing the mathematical convergence of environmental metrics against governance indices, the project provides a scalable framework to detect corporate "greenwashing" and regulatory decoupling. The results show that the model successfully isolates critical societal friction points—such as mapping structural safety deficits in the industrial/aerospace sectors or flagging high governance risk imbalances in the financial sector—proving how technical data pipelines can directly safeguard public trust and ensure sustainable capital allocation.
